```python
from __future__ import unicode_literals
import youtube_dl
import requests
from lxml import html
from bs4 import BeautifulSoup as bs
```

```python
#Scrape original mp3s from youtube
bts = [
    'fake love', 'dna', 'idol', 'boy with luv', 'dope', 'fire', 'i need u',
    'blood sweat and tears', 'not today', 'mic drop'
]
blackpink = [
    'kill this love', 'ddu-du ddu-du', 'boombayah', "as if it's your last",
    "don't know what to do", 'playing with fire', 'whistle',
    'forever young', 'kiss and make up', 'kick it'
]
dreamcatcher = [
    'chase me', 'good night', 'full moon', 'piri', 'you and i', 'fly high',
    'over the sky', 'what', 'wonderland', 'and there was no one left'
]
exo = [
    'love shot', 'ko ko bop', 'monster', 'tempo', 'growl', 'call me baby',
    'the eve', 'lotto', 'power', 'love me right'
]
monsta_x = [
    'trespass', 'x', 'play it cool', 'shoot out', 'dramarama', 'alligator',
    'hero', 'jealousy', 'beautiful', 'shine forever'
]
aoa = [
    'miniskirt', 'like a cat', 'excuse me', 'confused', 'bingle bangle',
    'ai wo choudai', 'good luck', 'heart attack', 'bing bing', 'get out'
]
songs_lst = [bts, exo, monsta_x, dreamcatcher, aoa, blackpink]
headers = {
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'
}
for i in range(len(songs_lst)):
    if i == 0:
        group = 'bts'
    elif i == 1:
        group = 'exo'
    elif i == 2:
        group = 'monsta_x'
    elif i == 3:
        group = 'dreamcatcher'
    elif i == 4:
        group = 'aoa'
    else:
        group = 'blackpink'
    for song in songs_lst[i]:
        full_search = group + ' ' + song + ' lyrics'
        query = 'https://www.youtube.com/results?search_query=' + full_search.replace(' ', '+')
        page = requests.get(query)
        tree = bs(page.text, 'html.parser')
        vids = tree.findAll('a',attrs={'class':'yt-uix-tile-link'})
        url = 'https://www.youtube.com' + vids[0]['href']
        if 'googleadservices' in url: #first result might be an ad
            url = 'https://www.youtube.com'+ vids[1]['href']
        outtmpl = 'audio_files/' + group + '/'+ song.replace(' ', '_') + '.%(ext)s'
        ydl_opts = {
            'format': 'bestaudio/best',
            'outtmpl': outtmpl,
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'mp3',
                'preferredquality': '192',
            }],
        }
        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
```
